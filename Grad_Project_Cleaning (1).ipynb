```python
product_counts = df['Product line'].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(product_counts.index, product_counts.values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b'])

plt.xlabel('Product Line', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Distribution of Product Line', fontsize=14)
plt.xticks(rotation=45, ha='right')  # تدوير أسماء المنتجات لتكون أوضح
plt.grid(axis='y', linestyle='--', alpha=0.7)  # إضافة خطوط أفقية للمساعدة في القراءة

plt.show()
```
